```python
import pandas as gropd
import numpy as np
import time
import sys
import matplotlib.pyplot as plt
import os

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
#nltk.download("wordnet")
lemmatizer = WordNetLemmatizer()
import re
```

```python
data_path = ""
encoding = "ISO-8859-1"
df = gropd.read_csv(data_path + "merged_data.csv", encoding=encoding)
```

```python
df = df[:50000]
df
```

```python
stopwords.words('english')

def tokenizer_cleaning(sentences):
    text_tokens = word_tokenize(sentences)
    tokens_without_sw = [lemmatizer.lemmatize(word) for word in text_tokens if not word in stopwords.words()]    

    return tokens_without_sw
```

```python
def col_tokenizer(col):
    df[col] = [re.sub('\W', ' ', str(row)).lower() for row in df[col]]
    df[col] = df[col].values.tolist()
    df[col] = [tokenizer_cleaning(body) for body in df[col]]
```

```python
col_tokenizer("Body_Q")
col_tokenizer("Body_A")
```

```python
import numpy as np
import pandas as pd
```

```python
data_path = "data/"
encoding = "ISO-8859-1"
```

```python
questions = pd.read_csv(data_path + "Questions.csv", encoding=encoding)
answers = pd.read_csv(data_path + "Answers.csv", encoding=encoding)
tags = pd.read_csv(data_path + "Tags.csv", encoding=encoding)
```

```python
questions.head()
```

```python
answers.head()
```

```python
tags.head()
```

```python
tags.Tag = tags.Tag.astype(str)
tags.groupby('Id')['Tag'].apply(', '.join).reset_index()
```

```python
qna_data = questions.merge(answers, how='inner', left_on='Id', right_on='ParentId', suffixes=('_Q','_A'))
qna_data.merge(tags.groupby('Id')['Tag'].apply(', '.join), how='inner', left_on='Id_Q', right_on='Id')
```

```python
qna_data.head()
```

```python
qna_data.to_csv('data/merged_data.csv')
```

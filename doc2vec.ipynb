```python
import tensorflow as tf
import numpy as np 
import tensorflow.compat.v1 as tf

#si vous avez un tf 2.x
tf.disable_v2_behavior() 


import pandas as pd
import re
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from ast import literal_eval

filename = './../enwiki_dbow/doc2vec.bin'
model = Doc2Vec.load(filename)
```

```python
converters={"Body_A": literal_eval, "Body_Q": literal_eval}
df = pd.read_csv('preprocessed_data.csv', converters=converters)
df = df[:25000]
df = df[["Body_Q", "Body_A"]]
df["A"] = np.array
df["Q"] = np.array
df_test = df[12500:]
df_train = df[:12500]
```

```python
for index, row in df_train.iterrows():
    df_train.at[index, 'A'] = model.infer_vector(row['Body_A'])
    df_train.at[index, 'Q'] = model.infer_vector(row['Body_Q'])

for index, row in df_test.iterrows():
    df_test.at[index, 'A'] = model.infer_vector(row['Body_A'])
    df_test.at[index, 'Q'] = model.infer_vector(row['Body_Q'])
```
